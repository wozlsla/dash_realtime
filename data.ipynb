```python
ids_to_drop = [
    "0002",
    "0003",
    "0009",
    "0015",
    "0016",
    "0018",
]
data_ids_drop = data[~data["사용자ID"].isin(ids_to_drop)]
data_ids_drop
```

```python
workdays = data_ids_drop[data_ids_drop["근무일명칭"] == "평일"]
workdays_null = workdays[workdays["출근"].isna() & workdays["퇴근"].isna()]
workdays_null
```

```python
# ID 별로 비교, 겹치는 날이 있는지 확인
unique_work_dates = workdays_null["근무일자"].unique()
dates_per_id = workdays_null.groupby("사용자ID")["근무일자"].unique()

print(unique_work_dates, len(unique_work_dates), "\n")
print(dates_per_id)
```

```python
dates_per_id.items()
```

```python
date_dict = {}

# 올리버(0017) 9월 출근 X
for id, dates in dates_per_id.items():
    if id != "0017":
        date_dict[id] = dates
```

```python
date_dict
```

```python
# 네오(0001) 기준
common_dates = set(date_dict["0001"])

for dates_id in date_dict.values():
    common_dates.intersection_update(dates_id)

common_dates

# "평일"으로 기록 됨
# 2023/09/28 : 추석 연휴
# 2023/09/29 : 추석
# 2023/10/02 : 추석 대체 공휴일
# 2023/10/09 : 한글날
# 2023/10/16 : 전원 재택
```

```python
combined_filtered_data = data[
    (
        (data["근무일명칭"] == "개천절")
        | (data["근무일명칭"] == "성탄절")
        | (data["근무일명칭"] == "신정")
    )
    & (data["사용자ID"] == "0005")
]

combined_filtered_data
```

```python
# 공휴일(Public Holiday) 제외
public_holidays = [
    "2023/09/28",
    "2023/09/29",
    "2023/10/02",
    "2023/10/03",
    "2023/10/09",
    "2023/12/25",
    "2024/01/01",
]

data_phs_drop = data[~data["근무일자"].isin(public_holidays)]

# 근무일수 + 주말 : 146
data_phs_drop["근무일자"].nunique()
```

```python
# 근무일수
data_holiday_drop = data_phs_drop[data_phs_drop["근무일명칭"] == "평일"]
data_holiday_drop["근무일자"].nunique()
```

```python
ids_to_drop = [
    "0002",
    "0003",
    "0009",
    "0015",
    "0016",
    "0018",
]

commute = data_holiday_drop[~data_holiday_drop["사용자ID"].isin(ids_to_drop)]
```
